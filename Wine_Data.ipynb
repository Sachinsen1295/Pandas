```python
wine_df.groupby("winery").apply(lambda x:x.title.iloc[0])
```

```python
q=wine_df.groupby(["country","province"]).apply(lambda x:x.loc[x.points.idxmax()])
q.head(1)

#we need group data first by - country then province
# and afterthat start by max values in that column
#loc to lock the column "points"
#in lambda x is wine_df
#idxmax means -  returns a Series with the index of the maximum value for each
# column. By specifying the column axis ( axis='columns' ), the idxmax() method
 # returns a Series with the index of the maximum value for each row.
```

```python
wine_df.groupby("country").apply(lambda x:x.loc[x.points.idxmax()])
```
